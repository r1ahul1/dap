```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 100  # number of patients
probabilities = [0.2, 0.4, 0.8]  # success probabilities
num_simulations = 10000  # number of simulations for histogram

plt.figure(figsize=(14, 8))

for i, p in enumerate(probabilities, 1):
    # Simulation
        simulated_data = np.random.binomial(n, p, num_simulations)

# Theoretical distribution
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

# Mean and variance
mean, var = binom.stats(n, p)

# Plot histogram of simulation
plt.subplot(2, 2, i)
plt.hist(simulated_data, bins=20, density=True, alpha=0.6, color='skyblue', label='Simulation')
plt.plot(x, pmf, 'r-', lw=2, label='Theoretical PMF')

plt.axvline(mean, color='g', linestyle='--', label=f'Mean = {mean:.2f}')
plt.axvline(mean + np.sqrt(var), color='purple', linestyle=':', label=f'Variance = {var:.2f}')
plt.axvline(mean - np.sqrt(var), color='purple', linestyle=':')

plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Correct Diagnoses')
plt.ylabel('Probability Density')
plt.legend()

plt.tight_layout()
plt.show()
```
